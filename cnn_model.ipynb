```python
import pandas as pd
import gc
from keras import optimizers, losses, activations, models
from keras.layers import Convolution2D, Dense, Input, Flatten, Dropout, MaxPooling2D, BatchNormalization, UpSampling2D, InputLayer
from sklearn.model_selection import train_test_split
from keras.models import Sequential, Model
from keras.datasets import cifar100
```

```python
import numpy as np
from keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
import timeit
from keras import backend as K
import pickle
```

```python
K.clear_session()
model = models.Sequential()
model.add(InputLayer(input_shape=(32, 32, 3)))
#block 1
model.add(Convolution2D(64, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
#block 2
model.add(Convolution2D(64, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.3))
#block 3
model.add(Convolution2D(128, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
#block 4
model.add(Convolution2D(128, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.3))
# block 5
model.add(Convolution2D(256, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
#block 6
model.add(Convolution2D(256, kernel_size=3,padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.3))
# last block
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(100, activation='softmax'))
model.summary()
```

```python
(x_train, y_train), (x_test, y_test) = cifar100.load_data(label_mode='fine')
```

```python
x_train = x_train.astype('float32')/225
x_test = x_test.astype('float32')/225
y_train = to_categorical(y_train,100)
y_test = to_categorical(y_test,100)
```

```python
x_train_mean = np.mean(x_train, axis=0)
x_train -= x_train_mean
x_test -= x_train_mean
```

```python
model.compile(loss=losses.categorical_crossentropy,optimizer=optimizers.Adadelta(),metrics=['accuracy'])
```

```python
start = timeit.default_timer()

history=model.fit(x_train, y_train,batch_size=256,epochs=50,verbose=2,validation_data=(x_test, y_test), shuffle=True)
model.save('cnn_cifar2')

#print time taken to run this cell
end = timeit.default_timer()
print("Time Taken to run the model:",end - start, "seconds")
```
